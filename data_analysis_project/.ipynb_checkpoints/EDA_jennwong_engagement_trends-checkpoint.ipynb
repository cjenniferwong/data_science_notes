## Engagement Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from glob import glob
import time
from datetime import datetime
```

```python
data_files = glob('../data/*')
data_files
```

```python
reactions_data = pd.read_csv(data_files[1])

reactions_data.head()
```

```python
reactions_data.info()
```

```python
# cleaning the data

# NTS there is a pandas pipe function that works analogous to sklearn to chain together methods

reactions_data.columns = [str.lower(column) for column in reactions_data.columns]
reactions_data.columns = [column.replace(' ', '_') for column in reactions_data.columns]

reactions_data.rename(columns ={
    'id':'reaction_id'
}, inplace=True)

reactions_data.created_at = pd.to_datetime(reactions_data.created_at)

# cast it as date since the time doesnt matter since its in UTC
reactions_data.created_at = reactions_data.created_at.dt.date

# figure out why this takes so long..
```

```python
# going to replace the two different hearts with just one to clean in


set(reactions_data.emoji)
```

```python
reactions_data.emoji = reactions_data.emoji.replace('‚ô•Ô∏è', '‚ù§Ô∏è')
```

```python
# need to set the time as the index to be grouped by the month
reactions_data.head()
```

```python
# a different way of analysis instead of group by

reactions_data.emoji.value_counts()
```

## typically we would want to look at the impressions too to get an idea of the whole picture but i dont have the impression data rn 

```python
reactions_data.set_index(pd.DatetimeIndex(reactions_data.created_at))\
.groupby(pd.Grouper(freq='M')).count()['reaction_id'].plot()

plt.title('Engagement Over Time')

plt.show();
```

```python
# i want to know what is responsible for the spike in reactions between jan and jul 2018 
change_point = '2018-01-01'
end_point = '2018-07-01'
spiked_data = reactions_data[(reactions_data.created_at >= datetime.strptime(change_point, '%Y-%m-%d').date()) & \
               (reactions_data.created_at < datetime.strptime(end_point, '%Y-%m-%d').date())
              ].sort_values('created_at', ascending=True)
```

```python
monthly_reaction_counts = spiked_data.set_index(pd.DatetimeIndex(spiked_data.created_at))\
.groupby([pd.Grouper(freq='M'), 'story']).count()['reaction_id']
```

```python
monthly_reaction_counts.head()
```

```python
spiked_data.set_index(pd.DatetimeIndex(spiked_data.created_at))\
.groupby([pd.Grouper(freq='M'), 'story']).count()
```
